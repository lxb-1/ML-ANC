```python
from scipy.io import wavfile
import sys
sys.path.append("../") 
import pysepm
import IPython
```

# Load Audio Files

```python
fs, clean_speech = wavfile.read('1_speech_16000_Hz.wav')
fs, noisy_speech = wavfile.read('1_noisySpeech_16000_Hz.wav')
fs, enhanced_speech = wavfile.read('1_processed_16000_Hz.wav')
```

```python
print("清洗过了语音：")
IPython.display.Audio(clean_speech, rate=fs)
```

```python
print("带有噪声的语音：")
IPython.display.Audio(noisy_speech, rate=fs)
```

```python
print("模糊增强后的语音：")
IPython.display.Audio(enhanced_speech, rate=fs)
```

# Calc Measures

## fwSNRseg（频率加权分段信噪比）

```python
pysepm.fwSNRseg(clean_speech, noisy_speech, fs)
```

```python
pysepm.fwSNRseg(clean_speech, enhanced_speech, fs)
```
